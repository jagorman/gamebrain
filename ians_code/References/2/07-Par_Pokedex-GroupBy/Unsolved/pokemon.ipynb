```python
# Dependencies
import pandas as pd
```

```python
# Save file path to variable
file = "Resources/Pokemon.csv"
```

```python
# Read with Pandas
poke_df = pd.read_csv(file)
```

```python
poke_df
new_df = poke_df[["Name", "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
#This sets a new dataframe with the below extractions
poke_df
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
#Another way to create the new dataframe by modifying existing
del poke_df["Type 2"]
del poke_df["Generation"]
del poke_df["#"]
del poke_df["Legendary"]
del poke_df["Total"]
poke_df.head(3)
```

```python
new_df.head(3)
```

```python
# Create a dataframe of the average stats for each type of pokemon.
poke_group = new_df.groupby(["Type 1"])

poke_stats_df = poke_group.mean()
poke_stats_df
```

```python
# Calculate the total power level of each type of pokemon by summing all of the stats together.
# Place the results into a new column.
# Hint: Research the sum method from pandas.
poke_stats_df["Total Power"] = poke_stats_df.sum(axis=1)
#(axis=1) switches the sum from columns to rows
poke_stats_df
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
sorted_df = poke_stats_df.sort_values("Total Power",ascending=False)
#ascending=False puts strongest at the top
sorted_df
```

```python
sorted_df.to_csv("poke.csv")
```
