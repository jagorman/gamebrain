```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
file = 'Resources/crime_incident_data2017.csv'
# Import the data into a Pandas DataFrame
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df
```

```python
# look for missing values
df.count()
```

```python
# drop null rows
cleaned_df = df.dropna(how='any')
```

```python
# verify counts
cleaned_df.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
cleaned_df['Offense Type'].value_counts()
```

```python
# Combine similar offenses
#Real world you would clarify what would be grouped together before grouping
cleaned_df['Offense Type'] = cleaned_df['Offense Type'].replace(
    {'Theft From Motor Vehicle': 'Motor Vehicle Theft', 'Theft of Motor Vehicle Parts or Accessories': 'Motor Vehicle Theft'})
cleaned_df['Offense Type'] = cleaned_df['Offense Type'].replace(
    {'Robbery': 'Burglary', 'Shoplifting': 'Burglary','Pocket-Picking': 'Burglary','Purse-Snatching': 'Burglary','Theft From Building': 'Burglary',})
```
